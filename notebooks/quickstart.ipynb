# Starbox Quickstart

```python
from starbox.config import TelescopeConfig, SkyModelConfig
from starbox.factory import build_telescope, build_skymodel
from starbox.viz.plot import plot_telescope, plot_sky_model
```

## Simulate a small array of antennas

```python
telescope_config = TelescopeConfig(num_stations=10, diameter=35.0, seed=42)
telescope = build_telescope(telescope_config)
plot_telescope(telescope)
```

## Simulate a random sky model

```python
skymodel_config = SkyModelConfig(num_sources=20, seed=42, max_flux_jy=5.0, fov_deg=5.0)
skymodel = build_skymodel(skymodel_config)
plot_sky_model(skymodel)
```

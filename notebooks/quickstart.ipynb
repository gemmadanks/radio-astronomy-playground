# Starbox Quickstart

```python
from starbox.simulate import Telescope, SkyModel
from starbox.viz.plot import plot_array_configuration, plot_sky_model
```

## Simulate a small array of antennas

```python
telescope = Telescope(name="MyArray", num_stations=10, diameter=50.0)
plot_array_configuration(telescope)
```

## Simulate a random sky model

```python
skymodel = SkyModel(name="MySkyModel", num_sources=50, max_flux=10.0)
plot_sky_model(skymodel)
```
